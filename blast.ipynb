```python
# save as export_scn_fasta.py
import sidechainnet as scn
from pathlib import Path

out_fa = Path("sidechainnet_all.fasta").open("w")
out_fa_txt = "sidechainnet_all.fasta"
# CASP 7–13 are the public ProteinNet splits; 100 % thinning keeps *all* chains
for casp in range(12, 13):
    ds = scn.load(casp_version=casp, casp_thinning=100, scn_dataset=True)
    # write sequences in >id\nSEQ format
    ds.to_fasta(out_fa_txt)          # SCNDataset has this convenience fn (v1.0+) :contentReference[oaicite:1]{index=1}
    print(f"Wrote CASP{casp} ({len(ds)} proteins)")

out_fa.close()
```
